```python
import gensim.downloader as api
import numpy as np

print(f"NumPy version: {np.__version__}")
wv = api.load('word2vec-google-news-300')
```

```python
wv.similarity(w1='great',w2='good')
```

```python
wv.most_similar('good')
```

```python
wv.most_similar('dog')
```

```python
wv.most_similar(positive=['king','woman'],negative=['man'],topn=5)
```

```python
wv.most_similar(positive=['france','berlin'],negative=['paris'],topn=5)
```

```python
wv.doesnt_match(['facebook','cat','google','microsoft'])
```

```python
wv.doesnt_match(['dog','cat','google','mouse'])
```

```python
glv = api.load('glove-twitter-25')
```

```python
glv.most_similar('good')
```

```python
glv.doesnt_match('breakfast cereal dinner lunch'.split())
```

```python
glv.doesnt_match("banana grapes orange human".split())
```

```python
glv.doesnt_match('facebook cat google microsoft'.split())
```

##### Tools :

* Word2Vec
  focus : Neighbors: What words usually surround "Apple"?
  result :  *  Finds "ate," "crunchy," "juice."
  
* GloVe
  focus : Stats: How often does "Apple" appear
  result : Finds "iPhone," "Mac," "Tech." in the same book as "Steve Jobs"?

* Gensim
  focus : Action: The user wants to compare  "Apple" and "Banana."
  result : Runs the math and says: "They are 85% similar."
