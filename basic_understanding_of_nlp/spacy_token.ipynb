### Difference between NLTK and spacy

### NLTK :
* It is a string processing library.
* It is user friendly but probably less user friendly compared to spacy.
* Provides access to many algorithms.if you care about specific algo and customizations go with NLTK
* NLTK is old library.user community as active as spacy.

### Spacy
* It is object oriented
* it is user friendly
* Provides most efficient NLP algorithm for a given task.Hence if you care about the end result,go with Spacy
* it is a new library and has a very active user community


Customizing tokenizer

```python
from spacy.symbols import ORTH
nlp = spacy.blank('en')
doc = nlp("gimme double cheese extra large healthy pizza")
tokens = [token.text for token in doc]
tokens
```

```python
nlp.tokenizer.add_special_case('gimme',[{ORTH:'gim'},{ORTH:'me'}])
doc = nlp("gimme double cheese extra large healthy pizza")
tokens = [token.text for token in doc]
tokens
```

```python
doc = nlp("Captain america ate 100$ of samosa. Then he said I can do this all day.")
for token in doc:
  print(token," | ",spacy.explain(token.pos_)," | ",token.lemma_)
```


Named Entity Recognition

```python
doc = nlp("Tesla Inc is going to acquire twitter for $45 billion")
for ent in doc.ents:
  print(ent.text,ent.label_)
```

```python
from spacy import displacy
displacy.render(doc,style='ent')
```


Adding a component to a blank pipeline

```python
source_nlp = spacy.load("en_core_web_sm")
nlp = spacy.blank("en")
nlp.add_pipe("ner",source=source_nlp)
nlp.pipe_names
```

```python
doc = nlp("Tesla Inc is going to acquire twitter for $45 billion")
for ent in doc.ents:
  print(ent.text,ent.label_)
```
