```python
def get_pattern_match(pattern,text):
  matches = re.findall(pattern,text)
  if(matches):
    return matches[0]
```

```python
customer_reviews = ['sam was a great help to me in the store',
                    'the cashier was very rude to me, I think her name was eleanor',
                    'amazing work from sadeen!',
                    'sarah was able to help me find the items i needed quickly',
                    'lucy is such a great addition to the team',
                    'great service from sara she found me what i wanted'
                   ]
```

```python
sara_review = []
new_search = r'sarah?'
```

```python
for string in customer_reviews:
  if re.search(new_search,string):
    sara_review.append(string)
```

```python
print(sara_review)
```

```python
a_review = []
a_search = r'^a'
```

```python
for i in customer_reviews:
  if re.search(a_search,i):
    a_review.append(i)
```

```python
print(a_review)
```

```python
y_review = []
y_search = r'y$'
```

```python
for i in customer_reviews:
  if re.search(y_search,i):
    y_review.append(i)
```

```python
print(y_review)
```

```python
nw_review = []
nw_search = r'(need|want)ed'
```

```python
for i in customer_reviews:
  if re.search(nw_search,i):
    nw_review.append(i)
```

```python
print(nw_review)
```

```python
pattern_review = []
pattern_to_find = r'[^\w\s]'
```

```python
for i in customer_reviews:
  finding = re.sub(pattern_to_find,"",i)
  pattern_review.append(finding)
```

```python
print(pattern_review)
```
